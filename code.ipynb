
TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
  x=asd_data.drop(['Class','Who_completed_the_test'],axis=1)
  y=asd_data['Class']
  x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=101)
  #print(x_train)
  #print(x_test)
  x_train.head(10)
```

```python
!mkdir Training_Testing2
```

```python
x_train.to_excel('Training_Testing2/x_train.xlsx',index=False)
x_test.to_excel('Training_Testing2/x_test.xlsx',index=False)
y_train.to_excel('Training_Testing2/y_train.xlsx',index=False)
y_test.to_excel('Training_Testing2/y_test.xlsx',index=False)
```

```python
class_labels=['No','Yes']
```

TRAINING THE RANDOM FOREST MODEL

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
rfc=RandomForestClassifier(n_estimators=600)
```

```python
rfc=rfc.fit(x_train,y_train)
```

PREDICTIONS AND EVALUATIONS OF RF MODEL

```python
predictions_rf=rfc.predict(x_test)
```

```python
acc_rf=accuracy_score(y_true=y_test,y_pred=predictions_rf)
print("%f"%(acc_rf*100))
```

```python
print(classification_report(y_test,predictions_rf,target_names=class_labels))
```

```python
plt.figure(figsize=(5,5))
plt.rcParams['font.size']=15
sns.heatmap(data=confusion_matrix(y_true=y_test,y_pred=predictions_rf),
            annot=True,
            fmt='4d',
            cbar=False,
            cmap=plt.cm.Blues,
            xticklabels=class_labels,
            yticklabels=class_labels)
plt.title(label='confusion matrix')
plt.yticks(rotation=0)
plt.show
```

TRAINING THE ADABOOST MODEL

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
ada=AdaBoostClassifier(n_estimators=500)
```

```python
ada.fit(x_train,y_train)
```

PREDICTION AND CLASSIFICATION OF ADABOOST

```python
predictions_ada=ada.predict(x_test)
```

```python
acc_ada=accuracy_score(y_true=y_test,y_pred=predictions_ada)
print("%f"%(acc_ada*100))
```

```python
print(classification_report(y_test,predictions_ada,target_names=class_labels))
```

```python
plt.figure(figsize=(5,5))
plt.rcParams['font.size']=15
sns.heatmap(data=confusion_matrix(y_true=y_test,y_pred=predictions_ada),
            annot=True,
            fmt='4d',
            cbar=False,
            cmap=plt.cm.Blues,
            xticklabels=class_labels,
            yticklabels=class_labels)
plt.title(label='confusion matrix')
plt.yticks(rotation=0)
plt.show
```
